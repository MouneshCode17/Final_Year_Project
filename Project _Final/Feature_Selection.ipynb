```python
import pandas as pd
```

```python
df = pd.read_csv('IPL_CSV.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Assuming df is your DataFrame containing cricket match data

label_encoder_players = LabelEncoder()

# Combine 'batter', 'bowler', and 'non-striker' into a single Series
combined_categories = pd.concat([df['batter'], df['bowler'], df['non-striker'],df['Player_of_Match'],df['fielders_involved']], ignore_index=True)

# Fit label encoder on combined categories
label_encoder_players.fit(combined_categories)

# Apply label encoding to each column with handling unseen labels
df['batter'] = label_encoder_players.transform(df['batter'])
df['bowler'] = label_encoder_players.transform(df['bowler'])
df['non-striker'] = label_encoder_players.transform(df['non-striker'])
df['Player_of_Match']= label_encoder_players.transform(df['Player_of_Match'])
df['fielders_involved'] = label_encoder_players.transform(df['fielders_involved'])
```

```python
df
```

```python
label_encoder_teams = LabelEncoder()

# Combine 'batter', 'bowler', and 'non-striker' into a single Series
combined_categories = pd.concat([df['Team1'], df['Team2'], df['BattingTeam'],df['BowlingTeam'],df['WinningTeam'],df['TossWinner']], ignore_index=True)


label_encoder_teams.fit(combined_categories)

# Apply label encoding to each column with handling unseen labels
df['Team1'] = label_encoder_teams.transform(df['Team1'])
df['Team2'] = label_encoder_teams.transform(df['Team2'])
df['BattingTeam'] = label_encoder_teams.transform(df['BattingTeam'])
df['BowlingTeam'] = label_encoder_teams.transform(df['BowlingTeam'])
df['WinningTeam'] = label_encoder_teams.transform(df['WinningTeam'])
df['TossWinner'] = label_encoder_teams.transform(df['TossWinner'])
```

```python
df['crr'] = df['crr'].abs()
df['rrr'] = df['rrr'].abs()
df['runs_left'] = df['runs_left'].abs()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder_rem = LabelEncoder()

# Fit label encoder and transform 'City' column
df['City'] = label_encoder_rem.fit_transform(df['City'])

df['TossDecision'] = label_encoder_rem.fit_transform(df['TossDecision'])

df['kind'] = label_encoder_rem.fit_transform(df['kind'])

df['WonBy'] = label_encoder_rem.fit_transform(df['WonBy'])

df['extra_type'] = label_encoder_rem.fit_transform(df['extra_type'])
```
